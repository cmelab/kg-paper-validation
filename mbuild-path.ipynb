````python
import mbuild as mb
from mbuild.path import HardSphereRandomWalk
from mbuild import Polymer

import numpy as np
from scipy.spatial.distance import pdist
import freud
import matplotlib.pyplot as plt
import time
````

````python
rw_path = HardSphereRandomWalk(
    N=80,
    bond_length=0.25 * 1.15,
    radius=0.25,
    min_angle=np.pi/2,
    max_angle=2.5,
    max_attempts=1e4,
    seed=24,
)
rw_path.generate()
````

````python
comp = rw_path.to_compound()
comp.visualize(bead_size=0.8)
````

````python
chain = Polymer()
ethane = mb.load("CC", smiles=True)
chain.add_monomer(compound=ethane, indices=[2, 6], replace=True, separation=0.154)
chain.build_from_path(path=rw_path, energy_minimize=True)
chain.visualize()
````

````python
n_chains = 30
chain_lengths = 40
seeds = [np.random.randint(low=1) for i in range(n_chains)]
chains = []
for seed in seeds:
    rw_path = HardSphereRandomWalk(
        N=chain_lengths,
        bond_length=0.25 * 1.15,
        radius=0.25,
        min_angle=np.pi/2,
        max_angle=2.5,
        max_attempts=1e4,
        seed=seed,
    )
    rw_path.generate()

    chain = Polymer()
    ethane = mb.load("CC", smiles=True)
    chain.add_monomer(compound=ethane, indices=[2, 6], replace=True, separation=0.154)
    chain.build_from_path(path=rw_path, energy_minimize=True)
    chains.append(chain)
````

````python
chain_length = 50
n_chains = 16

seeds = [np.random.randint(low=1) for i in range(n_chains)]
chains = []

for seed in seeds:
    rw_path = HardSphereRandomWalk(
        N=chain_length,
        bond_length=1.12,
        radius=1.1,
        min_angle=np.pi/2,
        max_angle=2.5,
        max_attempts=1e4,
        seed=seed,
    )
    rw_path.generate()
    chains.append(rw_path.to_compound(bead_name="A", bead_mass=1.0))

box = mb.fill_box(compound=chains, n_compounds=[1 for i in chains], box=[50,50,50],overlap=1.1)
box.check_for_overlap(minimum_distance=1.0, excluded_bond_depth=1)
box.visualize(bead_size=0.8)
````

````python
box.to_hoomdsnapshot()
````
